### OpenCV by default reads channels as BGR

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Scale image to grayscale and check
img_gray = cv2.imread('images/rock.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
img_gray.shape
```

```python
# Resize image
img2 = cv2.resize(img,(100,50))
plt.imshow(img2)
```

```python
# Resize image with ratio
width_ratio = 0.5
height_ratio = 0.5
img3 = cv2.resize(img,(0,0),img,width_ratio,height_ratio)
plt.imshow(img3)
img3.shape
```

```python
# Flip on horizontal axis
img4 = cv2.flip(img,0)
plt.imshow(img4)
```

```python
# Flip on vertical axis
img5 = cv2.flip(img,1)
plt.imshow(img5)
```

```python
# Flip on both horizontal axis and vertical axis
img6 = cv2.flip(img,-1)
plt.imshow(img6)
```

```python
# Change the shape of our canva
last_image = plt.figure(figsize=(10,7))
ilp = last_image.add_subplot(111)
ilp.imshow(img)
```

# Drawing shapes


```python
#Create a black image
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
# Get shape of image
print(black_img.shape)
# Show the image
plt.imshow(black_img)
```

```python
# draw a circle
cv2.circle(img=black_img,center=(400,100),radius=50,color = (255,0,0),thickness=5)
plt.imshow(black_img)
```

```python
# draw a filled circle
cv2.circle(img=black_img,center=(400,300),radius=50,color = (0,255,0),thickness=-1)
plt.imshow(black_img)
```

```python
# draw a rectangle
cv2.rectangle(black_img,pt1=(50,50),pt2=(150,150),color = (0,255,0),thickness=4)
plt.imshow(black_img)
```

```python
# draw a filled rectangle
cv2.rectangle(black_img,pt1=(50,200),pt2=(150,300),color = (255,0,0),thickness=-1)
plt.imshow(black_img)
```

```python
# draw a triangle
vertices = np.array([[10,250],[110,150],[180,250]],np.int32)
pts = vertices.reshape(-1,1,2)
cv2.polylines(black_img,pts,isClosed=True,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
# draw a filled triangle
vertices = np.array([[10,250],[110,150],[180,250]],np.int32)
pts = vertices.reshape(-1,1,2)
cv2.fillPoly(black_img,pts,color=(0,0,255))
plt.imshow(black_img)
```

```python
# draw a line
cv2.line(black_img,pt1=(0,0),pt2=(512,512),color=(0,0,255),thickness=8)
plt.imshow(black_img)
```

```python
# write a text
msg = "Hi"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img,text=msg,org=(200,300),fontFace=font,fontScale=3,color=(255,255,0),thickness=8,lineType=cv2.LINE_AA)
plt.imshow(black_img)
```

# Drawing with Mouse

```python
# Function
# x,y,flags,param are feed from OpenCV automatically
def draw_circle(event, x, y,flags,param):
  if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),70,(35,69,78),1)
# Connect the function with call back
cv2.namedWindow(winname = 'Sketch')
# Callback
cv2.setMouseCallback('Sketch',draw_circle)
#Using OpenCV to show the image
img = np.zeros((512,512,3),np.int8)
while True:
  cv2.imshow('Sketch',img)
  if cv2.waitKey(5) & 0xFF == 27:
    break
cv2.destroyAllWindows()
```

# Add choice to mouse action

```python
# Function

# x,y,flags,param are feed from OpenCV automatically
def draw_circle(event, x, y,flags,param):
  if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),70,(75,131,255),1)
  elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),50,(255,75,131),1)

# Connect the function with call back
cv2.namedWindow(winname = 'Sketch')

# Callback
cv2.setMouseCallback('Sketch',draw_circle)

#Using OpenCV to show the image
img = np.zeros((512,512,3),np.int8)
while True:
  cv2.imshow('Sketch',img)
  if cv2.waitKey(5) & 0xFF == ord('q'):
    break
cv2.destroyAllWindows()
```

# Add functionality to Mouse Action

```python
drawing = False
ex = -1
ey = -1

# Function
# x,y,flags,param are feed from OpenCV automatically
def draw_circle(event, x, y,flags,param):
  global ex, ey, drawing
  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    ex,ey =x,y
  elif event == cv2.EVENT_MOUSEMOVE:
    drawing = True
    cv2.rectangle(img),(ex,ey),(x,y),(255,0,255),-1)
  elif event == cv2.EVENT_LBUTTONUP:
    drawing = False
    cv2.rectangle(img),(ex,ey),(x,y),(255,0,255),-1)

# Connect the function with call back
cv2.namedWindow(winname = 'Sketch')

# Callback
cv2.setMouseCallback('Sketch',draw_circle)

#Using OpenCV to show the image
img = np.zeros((512,512,3),np.int8)
while True:
  cv2.imshow('Sketch',img)
  if cv2.waitKey(5) & 0xFF == ord('q'):
    break
cv2.destroyAllWindows()


```
